```python
#helper function for preparing directories. do not change this.

def make_path(*argv):
    mypath = path.join(*argv)
    if not path.exists(dirname(mypath)):
        os.mkdir(dirname(mypath))
    return mypath

def make_dir(*argv):
    mydir = path.join(*argv)
    if not path.exists(mydir):
        os.mkdir(mydir)
    return mydir
```

```python
# protein_id = 'XP_003712653.1'
protein_id = 'XP_003711083.1'
genus = 'magnaporthe'

entropy_seq_path = make_path('entropy', genus, 'seq', 'entropy_clustalo_%s.npy' % protein_id)    
entropies = np.load(entropy_seq_path)
```

```python
-(list(enumerate(entropies))[10][1])
```

```python
attr = 'attribute: percentExposed\nrecipient: residues\n'
for ii,bit in enumerate(entropies):    
    attr += '\t:%d\t%.2f\n' %(ii+1,abs(bit))

attr_path = make_path('attr', genus, protein_id + '.txt')
with open(attr_path, 'w') as f:
    f.write(attr)
```
