```python
#helper function for preparing directories. do not change this.

def make_path(*argv):
    mypath = path.join(*argv)
    if not path.exists(dirname(mypath)):
        os.mkdir(dirname(mypath))
    return mypath

def make_dir(*argv):
    mydir = path.join(*argv)
    if not path.exists(mydir):
        os.mkdir(mydir)
    return mydir
```

```python
ref_protein = next(SeqIO.parse('nlr_protein/magnaporthe/nlr_XP_003712653.1.fasta', 'fasta'))
```

```python
# start = 865
# end = 1115
window_size = 15
matrix = matlist.blosum62

window_start = 997
window = ref_protein[window_start:window_start +window_size]
```

```python
print(window)
```

```python
def align(record1, record2):
    return pairwise2.format_alignment(*pairwise2.align.globalds(record1, record2, matrix, -window_size*10, -window_size*10)[0])
```

```python
print(align(window.seq, 'GHALQAASYXXXYMEIVELLLEKGADVNAQGGK'[:window_size]))
```

```python
protein_id = 'XP_003712653.1'
```

```python
threshold = 13
anks_dict = {}
rep_lens = []
last_rep_lens = []
for protein in SeqIO.parse('exonerate_p2gaa/magnaporthe/p2gaa_%s.faa'%protein_id, 'fasta'):
#     if protein.id=='PJXQ01002869.1':
    anks_dict[protein.id] = []
    last_spike = 0
    protein.seq = str(protein.seq).replace('*', '')
    scores = []
    ankNum = 1
    start = 0
    for jj in range(start, len(protein) - window_size):
        curr_score = pairwise2.align.globalds(window.seq, protein[jj: jj+window_size].seq, matrix, -window_size*10, -window_size*10, score_only = True, one_alignment_only = True)
        if curr_score > threshold:
            
            if not last_spike:                                                    
                last_spike = jj
            elif jj - last_spike > 15:    
                if jj - last_spike < 36:                    
                    ank = protein[last_spike: jj-1]
                    print(len(ank))
                    print(ank)
                    ank.id += '.%d'%ankNum
                    ank.name += '.%d'%ankNum
                    ankNum += 1
                    anks_dict[protein.id].append(ank)
                    rep_lens.append(jj - last_spike)
                    
                last_spike = jj
            else:
                curr_score = 0
        else:
            curr_score = 0

        scores.append(curr_score)

    
    last_rep = protein[last_spike:]
    if window_size < len(last_rep) < 36 and last_spike:
        if len(last_rep)>33:
            last_rep = last_rep[:33]
        last_rep.id += '.%d'%ankNum
        last_rep.name += '.%d'%ankNum
#         print(last_rep.seq)
        anks_dict[protein.id].append(last_rep)
        last_rep_lens.append(len(last_rep))

#             plt.close()
#             plt.plot([score*(score>threshold) for score in scores])
#             plt.xlim(700, len(protein))
#             plt.title(protein.id)
#             plt.show()

```
